```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
s=Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=s)
driver.get("https://www.zumper.com/rent-research/toronto-on")
```

```python
from selenium.webdriver.support.ui import Select
import time
```

```python
def get_data():
    output = {}
    try:
        select = Select(driver.find_element(By.XPATH,'//select[@aria-label="bedroom"]'))
    except:
        return None
    output["market_summary"] = {}
    for i in range(5):
        select.select_by_value(str(i))
        time.sleep(0.1)
        output["market_summary"][str(i)] = [i.text.split("\n")[0] for i in driver.find_elements(By.CLASS_NAME,'About_infoText__1x5Md')]
    output["rent_ranges"] = [i.get_attribute("ct:value")[:4] for i in driver.find_elements(By.CLASS_NAME,'ct-bar')]
    time.sleep(2)
    table_button = driver.find_element(By.XPATH,'//button[@id="type-tabs--tab-1"]')
    driver.execute_script("arguments[0].click();", table_button)
    output["historical_data"] = {}
    for k in range(5):
        try:
            tabs = driver.find_element(By.XPATH,'//div[@role="tablist"]').find_elements(By.TAG_NAME, "button")
            driver.execute_script("arguments[0].click();", tabs[k])
            time.sleep(0.1)
            table = [i for i in driver.find_elements(By.TAG_NAME,'tbody') if "PriceTable_tableBody" in i.get_attribute("class")][0]
            table_data = [i.text for i in table.find_elements(By.TAG_NAME, "td")]
            temp = {}
            for i in range(len(table_data) // 2):
                temp[table_data[i * 2]] = table_data[i * 2 + 1]
            output["historical_data"][str(k)] = temp
        except:
            pass
    return output
```
